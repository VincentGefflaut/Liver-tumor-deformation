```python
import numpy as np
import nibabel as nib
# import nilearn as nl
import nilearn.plotting as nplt
import os
import napari
import matplotlib.pyplot as plt
```

```python
img_path = "data/Volumes foies ablation anonymisés/Patient test 2_segmentation/Patient test 2-liver_post_ablation.nii.gz"

img = nib.load(img_path)

print(f"Dimensions : {img.shape}")

# 2. Résolution des voxels en mm
print(f"Résolution (mm) : {img.header.get_zooms()}")

# 3. Type de données
print(f"Type de données : {img.get_data_dtype()}")
```

```python
viewer = napari.view_image(img.get_fdata(), scale=img.header.get_zooms())
napari.run()
```

```python
orientation = nib.aff2axcodes(img.affine)
print(f"Orientation : {orientation}") 
# Exemple : ('L', 'A', 'S') pour Left, Anterior, Superior
```

```python
# 1. Charger l'image du patient
img_patient = nib.load('data/Volumes foies ablation anonymisés/Patient test 1_segmentations/Patient test 1-liver post ablation.nii.gz')
data_patient = img_patient.get_fdata()

# 2. Charger le masque de la zone d'ablation
img_mask = nib.load('data/Volumes foies ablation anonymisés/Patient test 1_segmentations/Patient test 1-ablation zone.nii.gz')
data_mask = img_mask.get_fdata()

# 3. Lancer le viewer
viewer = napari.Viewer()

# Ajouter l'image (en ajustant le contraste automatiquement)
viewer.add_image(data_patient, name='Patient', colormap='gray')

# Ajouter le masque (en le mettant en rouge transparent)
viewer.add_labels(data_mask.astype(int), name='Zone Ablation', opacity=0.5)

napari.run()
```

```python
print("Stats Patient:", np.min(data_patient), np.max(data_patient), np.unique(data_patient).size)
print("Stats Masque:", np.min(data_mask), np.max(data_mask), np.unique(data_mask).size)
```
